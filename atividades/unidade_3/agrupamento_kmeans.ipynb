```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2lab
from sklearn.cluster import KMeans
print(os.getcwd())
print(os.path.exists('leish_1_0.png'))
```

```python
original = imread(r'd:\unidade_3-20260523T123604Z-3-001\unidade_3\leish_1_0.png')
lab = rgb2lab(original)
#gray = original[:,:,0]
gray = lab[:,:,2]
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
original.shape
```

```python
k = 20
km = KMeans(n_clusters=k,n_init=2)
X = np.reshape(gray,(-1,1))
```

```python
X
```

```python
km.fit_predict(X)
```

```python
X.shape
```

```python
result = km.fit(np.reshape(gray,(-1,1)))
#result = km.fit(np.reshape(gray.ravel(),(-1,1)))
```

```python
final = np.reshape(result.labels_,gray.shape)
```

```python
for i in range(k):
    im1 = final == i
    plt.imshow(im1,cmap='gray')
    #plt.imshow(final)
    plt.show()
```
