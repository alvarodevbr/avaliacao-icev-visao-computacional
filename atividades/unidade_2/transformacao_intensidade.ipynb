```python
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np


import warnings
warnings.filterwarnings("ignore")
```

```python
def alargamento(imagem,k,e):

    new_image = 1/(1+(k/imagem)**e)
    
    return new_image
```

```python
def potencia(imagem,c,y):
    new_image = c*imagem**y
    return new_image
```

```python
def logaritmica(imagem,c):
    return c*np.log(1+imagem)
```

```python
r = np.arange(0,256)
```

```python
s = alargamento(r,125,10)*255
```

```python
plt.plot(s)
plt.show()
```

```python
im = imread('../unidade_1/histograma/picos.jpg')[:,:,0]
new_im = alargamento(im,190,2)
```
