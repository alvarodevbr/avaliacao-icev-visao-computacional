#### Processamento de Histograma:

- Um histograma pode ser visto como uma função de distribuição de frequência ou como uma função de distribuição de probabilidade;

$$p(r_k)=\frac{n_k}{n}.$$

- De maneira geral dizemos que $p(r_k)$ dá uma estimativa da probabilidade de ocorrência do nível de cinza $r_k$ na imagem.

#### Propriedade do histograma:

- As informações espaciais não são representadas;

- Um histograma é único para uma determinada imagem, mas o inverso não é verdadeiro;

- A movimentação de objetos em uma imagem não tem qualquer efeito sobre o seu histograma.

## Equalização de histograma

- É uma transformação dos níveis de cinza de uma imagem que visa aumentar o intervalo dinâmico melhorando o contraste de imagens adquiridas sob péssimas condições de iluminação;
- É útil para comparar cenas que foram adquiridas com iluminação diferente (normaliza a imagem);
- De modo geral o que se procura é obter um histograma plano.

```python
from skimage.exposure import equalize_hist
```

## Implementar os atributos acima em uma imagem.

Os seguintes valores são calculado com o histograma e são largamente utilizados para representar imagens:
- Média
- Variância
- Skewness
- Kurtosis
- Energia
- Entropia



```python
from skimage.exposure import histogram

def descritor_histograma(imagem):
    """Atributos de 1ª ordem calculados a partir do histograma normalizado (256 níveis)."""
    if imagem.ndim > 2:
        imagem = rgb2gray(imagem)

    hist, _ = histogram(imagem, nbins=256)
    p = hist.astype(float)
    p /= p.sum()

    niveis = np.arange(len(p))
    media = np.sum(niveis * p)
    var = np.sum((niveis - media) ** 2 * p)
    desvio = np.sqrt(var) if var > 0 else 1e-10

    sk = np.sum(((niveis - media) / desvio) ** 3 * p)
    kur = np.sum(((niveis - media) / desvio) ** 4 * p)
    ener = np.sum(p ** 2)
    p_nz = p[p > 0]
    entr = -np.sum(p_nz * np.log2(p_nz))

    return [media, var, sk, kur, ener, entr]

nomes_atributos = ['media', 'variancia', 'skewness', 'kurtosis', 'energia', 'entropia']
for nome, valor in zip(nomes_atributos, descritor_histograma(claro)):
    print(f'{nome}: {valor:.4f}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier as mlp
from sklearn.metrics import accuracy_score

def classification(X,y):

    """
    X é uma matriz onde cada linha corresponde aos atributos calculados com o descritor
    y são os rótulos da classe de cada imagem. Exemplo [0,1,0,0,1,2,2,0,0,0,...,1]
    """
    
    # separar conjuntos de treino e teste
    X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7)
    # criar instancia do classificador
    clf = mlp()
    # treinar o classificador
    clf.fit(X_train,y_train)
    # realizar a predicao
    pred = clf.predict(X_test)
    # calcular o resultado utilizando a acuracia
    acuracia = accuracy_score(y_test,pred)
    
    return acuracia
```

```python
from glob import glob
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Base RIMONE: retinopatia glaucomatosa (benigno x maligno)
base = Path('../../unidade_3/RIMONE-db-r2')
benigno_paths = sorted(glob(str(base / 'benigno' / '*.jpg')))
maligno_paths = sorted(glob(str(base / 'maligno' / '*.jpg')))

paths = benigno_paths + maligno_paths
y = np.array([0] * len(benigno_paths) + [1] * len(maligno_paths))
rotulos = {0: 'benigno', 1: 'maligno'}

print(f'Imagens: {len(paths)} ({len(benigno_paths)} benignas, {len(maligno_paths)} malignas)')

X = np.array([descritor_histograma(imread(p)) for p in paths])
acuracia = classification(X, y)
print(f'Acurácia (MLP, 70% treino): {acuracia:.4f}')

# Scatter plots dois a dois (atributos nas linhas, eixo x na coluna)
n = len(nomes_atributos)
fig, axes = plt.subplots(n, n, figsize=(14, 14))
cores = ['C0', 'C1']
for i in range(n):
    for j in range(n):
        ax = axes[i, j]
        if i == j:
            ax.hist(X[:, i], bins=25, color='gray', alpha=0.8)
        else:
            for label, cor in enumerate(cores):
                mask = y == label
                ax.scatter(X[mask, j], X[mask, i], s=10, alpha=0.45, c=cor, label=rotulos[label])
        if i == n - 1:
            ax.set_xlabel(nomes_atributos[j], fontsize=8)
        if j == 0:
            ax.set_ylabel(nomes_atributos[i], fontsize=8)
handles, labels_leg = axes[0, 1].get_legend_handles_labels()
fig.legend(handles, labels_leg, loc='upper right')
fig.suptitle('Atributos de histograma — pares de variáveis', y=1.01)
plt.tight_layout()
plt.show()

# Redução de dimensionalidade
X_pca = PCA(n_components=2, random_state=42).fit_transform(X)
X_tsne = TSNE(n_components=2, random_state=42, perplexity=30, init='pca').fit_transform(X)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
for ax, dados, titulo in [(ax1, X_pca, 'PCA'), (ax2, X_tsne, 't-SNE')]:
    for label, cor in enumerate(cores):
        mask = y == label
        ax.scatter(dados[mask, 0], dados[mask, 1], s=18, alpha=0.6, c=cor, label=rotulos[label])
    ax.set_title(titulo)
    ax.legend()
    ax.set_xlabel('componente 1')
    ax.set_ylabel('componente 2')
plt.tight_layout()
plt.show()
```
