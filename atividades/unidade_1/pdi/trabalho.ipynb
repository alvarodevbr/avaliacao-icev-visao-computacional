```python
from skimage.io import imread, imsave
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np

image = imread('/Users/alvaroaraujo/Documents/posia/visao-computacional/icev/codigos/unidade_1/cr7.jpg')
```

2) Leia uma imagem colorida, transforme a mesma em níveis de cinza. Após isso, atribua o valor zero a todos os pixels cujo vizinho à direita tenha o mesmo valor de intensidade. Salve o resultado em disco.

```python
gray.shape[1]
```

```python
li,co = gray.shape

for l in range(li):
    for c in range(0,co-1):
        if gray[l,c] == gray[l,c+1]:
            gray[l,c] = 0

plt.imshow(gray, cmap='gray')
```
