### Percorendo uma imagem

```python
linhas,colunas,dimensao = imagem.shape #pegar dimensoes da imagem
```

```python
#percorrer todas as linhas e colunas da imagem
#se o pixel for menor que 200, troque para verde
for l in range(linhas):
    for c in range(colunas):
        if (imagem[l,c,0] < 200):
            imagem[l,c,:] = [0,255,0]
```

## Atividade 
- Percorra sua imagem e ache o pixel com maior valor em cada uma das componentes

```python
img = imread('lenna.jpg')
```

```python
li,co,di = img.shape #pegar dimensoes da imagem
```

```python
# Inicializando os valores máximos
max_r = 0
max_g = 0
max_b = 0

# Inicializando as posições (caso a imagem seja toda preta, previne erros no print)
pos_max_r = [0, 0]
pos_max_g = [0, 0]
pos_max_b = [0, 0]

# Percorrer a imagem
for l in range(li):
    for c in range(co):
        
        # Canal R (Vermelho)
        if img[l, c, 0] > max_r:
            max_r = img[l, c, 0]
            pos_max_r = [l, c]
            
        # Canal G (Verde) - Corrigido de 'imagem' para 'img'
        if img[l, c, 1] > max_g:
            max_g = img[l, c, 1]
            pos_max_g = [l, c]
            
        # Canal B (Azul) - Corrigido de 'imagem' para 'img'
        if img[l, c, 2] > max_b:
            max_b = img[l, c, 2]
            pos_max_b = [l, c]

# Resultados
print(f"Máximo R: {max_r} | Posição: {pos_max_r}")
print(f"Máximo G: {max_g} | Posição: {pos_max_g}")
print(f"Máximo B: {max_b} | Posição: {pos_max_b}")
```
